```
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
```

```
url = "https://raw.githubusercontent.com/garvita-jain/Detecting-Contours/master/thumbs_up_down.jpg"
```

```
image = io.imread(url)
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image1 = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
cv2_imshow(image1)
```

```
_, binary = cv.threshold(image1, 225, 255, cv.THRESH_BINARY_INV)
cv2_imshow(binary)
```

```
contours, hierarchy = cv.findContours(binary, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
modified = cv.drawContours(image, contours, -1, (0,255,0), 3)
cv2_imshow(modified)
```
